```python
import matplotlib.pyplot as plt
```

```python
data_5_5 = {2: [4.88563723, 41.08410244], 
           12: [3.03248618, 41.42871194], 
           30: [2.87826066, 41.72649951], 
           48: [2.79132542, 41.9245029]}
data_10_10 = {2: [4.12493629, 36.66013506],
             12: [1.53712158, 43.05915301], 
             30: [2.70965302, 42.19630462], 
             48: [2.70527372, 41.39863761]}
data_25_25 = {2: [7.37981159, 51.42851966],
             12: [6.01367034, 47.67913259], 
             30: [7.05125523, 45.05826935], 
             48: [7.62087756, 45.69786464]}
data_50_50 = {2: [11.0637949296,63.7795166999],
             12: [11.9330800101, 50.7014775398], 
             30: [11.3779673301, 49.9683681498], 
             48: [14.4066890502, 51.6635030701]}
data = [data_5_5, data_10_10, data_25_25, data_50_50]
```

```python
def plot_datasets():
    frac_array = [5, 10, 25, 50]
    i = 0
    for data_set in data:
        plt.figure(figsize=(8,6))
        x = [key for key in data_set.keys()]
        y = [val[0] for val in data_set.values()]
        graph = plt.plot(x, y, color='grey', linestyle='dashed', marker='o', markerfacecolor='red', markersize=9)
        plt.xlabel('Core number')
        plt.ylabel('Average execution time in s')
        plt.title(f'Ops mix: {frac_array[i]} | {frac_array[i]} | {100 - 2*frac_array[i]}')
        i += 1;
        y = [val[1] for val in data_set.values()]
        graph = plt.plot(x, y, color='grey', linestyle='dashed', marker='o', markerfacecolor='orange', markersize=9)
        plt.legend(['Uniform distribution', 'Normal distribution'])
        plt.xticks([4, 12, 30, 48])
        plt.show()
```

```python
plot_datasets()
```
